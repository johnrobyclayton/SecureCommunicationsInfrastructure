```python
import pandas as pd
pd.set_option('display.max_colwidth',100)
data = pd.read_csv('SMSSpamCollection.tsv',sep='\t',header=None)
data.columns=['label','body_text']
data.head()
```

```python
import string
string.punctuation
punctuation_names={
    '!':' _<_exclamation_>_',
    '"':'_<_doublequote_>_',
    '#':'_<_hashmark_>_',
    '$':'_<_dollar_>_',
    '%':'_<_percent_>_',
    '&':'_<_ampersand_>_',
    "'":'_<_singlequote_>_',
    '(':'_<_opencurvedbracket_>_ ',
    ')':' _<_closedcurvedbracket_>_',
    '*':'_<_asterisk_>_',
    '+':'_<_plus_>_',
    ',':' _<_comma_>_',
    '-':'_<_dash_>_',
    '.':' _<_fullstop_>_',
    '/':'_<_forwardslash_>_',
    ':':'_<_colon_>_ ',
    ';':'_<_semicolon_>_ ',
    '<':'_<_openangledbracket_>_ ',
    '=':' _<_equal_>_ ',
    '>':' _<_closedangledbrancket_>_',
    '?':' _<_question_>_',
    '@':'_<_at_>_',
    '[':'_<_opensquarebrancket_>_ ',
    '\\':'_<_backslash_>_',
    ']':' _<_closedsquarebracket_>_',
    '^':'_<_caret_>_',
    '_':'_<_underscore_>_',
    '`':'_<_backtick_>_',
    '{':'_<_opencurlybracket_>_ ',
    '|':'_<_pipe_>_',
    '}':' _<_closedcurlybracket_>_',
    '~':'_<_tilde_>_',
}
#print(punctuation_names['"'])
def replace_punct(text):
    #[if char in string.puntuation [chaar for chaar in punctuation_names[char]] else char for char in text]
    returnlist=[]
    #return [[chaar for chaar in punctuation_names[char] if char in string.punctuation else char] for char in text]
    for char in text:
        if char in string.punctuation:
            for chaar in punctuation_names[char]:
                returnlist.append(chaar)
        else:
            returnlist.append(char)
            #print(returnlist)
            returnstring=""
            for x in returnlist:
                returnstring+=x
            #print(returnstring)
    return returnstring
def listToString(fromlist):
    new=""
    for x in fromlist:
        new+=x
    return new

#print(listToString(replace_punct('It\'s a long-way to old? london town.')))
def remove_punct(text):
    [char for char in text if char not in string.punctuation]
data['body_text_clean']=data['body_text'].apply(lambda x: replace_punct(x))
data.head()
```

```python
import re
def tokenize(text):
    tokens = re.split(' ',text)
    returntokens=[]
    for x in range(len(tokens)):
        #print(tokens[x])
        if re.match('^[A-Z].+$',tokens[x]) and not re.match('^[A-Z]*$',tokens[x])or re.match('^[A-Z]$',tokens[x]):
            returntokens.append('_<_capitalised_>_')
        elif re.match('^[A-Z]+$',tokens[x]):
            returntokens.append('_<_allcaps_>_')
            #tokens[x]='_<_capitalised_>_'+tokens[x].lower()
        #print(tokens[x])
        returntokens.append(tokens[x].lower())
    return returntokens

#print(tokenize('Its a long way to the Top.'))

data['body_text_tokenized']= data['body_text_clean'].apply(lambda x: tokenize(x))
data.head()
```

```python
import nltk
stopword=nltk.corpus.stopwords.words('english')
def remove_stopwords(tokenized_list):
    text=[word for word in tokenized_list if word not in stopword]
    return text
data['body_text_nostop']=data['body_text_tokenized'].apply(lambda x:remove_stopwords(x))
data['body_text_nostop']=data['body_text_tokenized']
data.head()
```
